```python
#Vẽ hình vuông sử dụng vòng lặp while

import turtle
pen = turtle.Turtle()
pen.pensize(5)
pen.hideturtle() # ẩn bút vẽ
do_dai = int(input("nhập vào độ dài cạnh hình vuông: "))
so_canh = 0
while so_canh < 4:
    # vẽ 1 cạnh hình vuông
    pen.forward(do_dai)
    pen.right(90)
    
    #vẽ các cạnh còn lại
    so_canh += 1

turtle.done()
    
```

```python
'''
        vẽ rùa tập thể dục
Viết chương trình để rùa chạy từ bên trái màn hình sang bên phải
lúc rùa chạy các nét vẽ không liền nhau, giá trị độ dài các nét vẽ tùy ý.
'''

# khai báo thư viên turtle
import turtle

# khai báo thư viện lấy độ dài các nét vẽ tùy ý
import random as r

# khởi tạo turtle
pen = turtle.Turtle()
pen.shape("turtle")

# Ẩn hình con rùa
pen.hideturtle()
pen.pensize(5)
pen.color("red")
pen.speed(2)
# nhấc bút vẽ
pen.penup()

#Đặt vị trí ban đầu của con rùa sang bên trái
# so với vị trí giữa màn hình 400 pixel
# Mục đích để rùa chạy không ra khỏi màn hình
# khi vòng lặp quá lớn
pen.goto(-400,0)

# hiển thị hình ảnh con rùa
pen.showturtle()

# Thực hiện cho rùa di chuyển từ trái sang phải
count = 0
while count < 10:
    # sinh 2 giá trị ngẫu nhiên
    down = r.randint(20,50)
    up = r.randint(20,50)
    # đặt bút vẽ
    pen.pendown()
    # rùa tiến về phía trước với giá trị ngẫu nhiên và không để lại bút vẽ
    pen.forward(down)
    
    pen.penup()
    # rùa tiến về phía trước với giá trị ngẫu nhiên và để lại bút vẽ
    pen.forward(up)
    count +=1

turtle.done()
```

```python
''' 
        vẽ rùa bị quản trong nhà
Vẽ vòng tròn thể hiện vòng giới hạn, không cho rùa chạy ra khỏi vòng này.
Vị trí ban đầu của rùa ở chính giữa vòng tròn, hướng rùa chạy ngẫu nhiên. 
Khi rùa chạy gặp phải vòng tròn thì bị bắt về lại đúng vị trí ban đầu, 
chính giữa vòng tròn, rùa tiếp tục cố gắng thoát khỏi vòng tròn theo hướng khác.
Cứ như thế, rùa không bao giờ thoát ra khỏi vòng tròn. 
Chương trình kết thúc khi số lần cố gắng trốn thoát của rùa = 10.
'''


#khai báo thư viện turtle
import turtle

# khai báo thư viện lấy số ngẫu nhiên
import random as r

# khởi tạo turtle
pen = turtle.Turtle()
pen.hideturtle() # ẩn con rùa

# đặt vị trí cho con trỏ
pen.penup()
pen.goto(0, -200)

# vẽ đường bao nhốt rùa(đường bao màu đen)
pen.pencolor("black")
pen.pensize(5)

pen.pendown()
pen.circle(200)

# đặt ví trí ban đầu cho con rùa (ở giữa vòng tròn bao)
pen.penup()
pen.pencolor('green')
pen.goto(0, 0)
pen.showturtle() #hiện con rùa ở giữa vòng tròn

# tạo hướng ngẫu nhiên cho rùa
huong = r.randint(0, 360)
pen.right(huong)

# cho rùa chạy thoát khỏi đường bao
# khi rùa chạy đến đường bao, bắt rùa quay về vị trí ban đầu(giữa đường bao)
so_lan_rua_co_gang_thoat = 0
while True:
    pen.speed(1)
    pen.forward(188)
    pen.goto(0, 0)
    huong = r.randint(0, 360)
    pen.right(huong)
    pen.showturtle()
    so_lan_rua_co_gang_thoat += 1
    
    if so_lan_rua_co_gang_thoat == 10:
        break


turtle.done()
```
